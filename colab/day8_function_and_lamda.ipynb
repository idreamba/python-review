## Function (ฟังก์ชัน)
ฟังก์ชัน คือ โปรแกรมย่อย sub program บางครั้งเรียกกว่า Method หรือ Procedure ใช้ทำงานที่ถูกเรียกใช้บ่อยๆ เพื่อลดคำสั่งในการเขียนโปรแกรมที่ซ้ำซ้อน ประเภทของฟังก์ชั่นจะมี 2 แบบ คือ

ฟังก์ชันแบบมาตรฐาน (Standard function)

และฟังก์ชันที่สร้างขึ้นมาเอง (User defined function)

```python
# สี่ผืนผ้า คือ กว้าง x ยาว
# x = int(input('กว้าง '))
# y = int(input('ยาว '))
# z = x * y
# print('พื้นที่สี่เหลี่ยมผืนผ้าคือ',z)

#เปลี่ยนงานที่ทำซ้ำๆ ให้เป็น function
def rectangle(width):
    result = width * width
    print('พื้นที่สี่เหลี่ยมคือ',result)

rectangle(10)
rectangle(2)
```

```python
# function bmi
# สูตร นำหนักตัว / (ส่วนสูง **2)

def bmi(weight,height):
    result = weight / (height/100)**2
    #print(result)
    return result

bmi(95,170)

# ถ้าต้องการเอาค่า bmi + 100
x = bmi(95,170)
print(x+100)
```
