```python
def enctypt_decrypt(k,txt,secret_key):
    txt_encrypt = ''
    temp = secret_key
    max = len(secret_key)-1
    idx = 0
    comp = 0
    try:
        for i in txt:
            if(i==' '):
                txt_encrypt += ' '
            for j in secret_key:
                idx = temp.index(j)+k
                if(i == j):
                    comp = max - temp.index(j)
                    if(idx <= max):
                        txt_encrypt += temp[temp.index(j)+k]
                    else:
                       txt_encrypt += temp[(k-1)-comp]
    except IndexError:
        print("Index out of range")
    return txt_encrypt

#ใช้ [::-1] เพื่อ reverse string
secret_key = 'abcdefghijklmnopqrstuvwxyz'[::-1]
txt = 'dvuklymbs'

print(secret_key)
print(txt)
print('----')
print(enctypt_decrypt(7,txt,secret_key))
```
