```python
#loop 5x5
loop = 5
for i in range(loop):
    if i%2 == 0:
        number = 2*i + 1
        for j in range(loop):
            if i+j >= 4:
                print(number, end='')
                number = number - 1
            else:
                print('-',end='')
        print()
    else:
        number = 1
        for j in range(loop):
            if i+j >= 4:
                print(number,end='')
                number = number + 1
            else:
                print('-', end='')
        print()
```
