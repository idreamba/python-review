# ข้อมูลชนิดสายอักษร (String)
ข้อมูลสายอักษร หรือสตริง คือ ข้อมูลที่เป็นตัวอักษร (Character) มาเรียงร้อยต่อกัน

การประกาศตัวแปรข้อมูลชนิดนี้ ต้องอยู่ภายใต้เครื่องหมาย '...' อัญประกาศเดียว (single quote) หรือ "..." (double quote) อัญประกาศคู่

```python
#ยกตัวอย่างเช่น
var1 = 'Hello World'
var2 = "Python basic"

print(var1);
print(var2);
```

```python
#การเข้าถึงตัวอักษรภายในชุดข้อมูล String
var1 = 'Hello World' #11 ตัว + \n = 12 ตัว

# (วงเล็บ), [ก้ามปู], {ปีกกา}
#ต้องการดึง e มาแสดงผล เราจะใช้ variable[ตำแหน่ง - 1]
print(var1[1])

#ให้ปริ้น e r l d คนละบรรทัด
print(var1[1])
print(var1[8])
print(var1[9])
print(var1[10])

#ให้ปริ้น e r l d ให้อยู่บรรทัดเดียวกัน
print(var1[1],var1[8],var1[9],var1[10])

#ปริ้นตัวเดียว
var3 = 'A'
print(var3[0])
```

```python
#การเข้าถึงแบบ [x:y] คั่นด้วยโคลอน ตัว x คือต้นทาง y คือปลายทาง ตำแหน่ง y + 1
var1 = 'Hello World'
pointer = var1[0:5]
print(pointer)
print(var1[6:11])

#ถ้าต้องการทั้งหมด
print(var1[0:])

#ถ้าต้องการแบบย้อนตัวอักษร จากทางขวาเริ่มที่ -1
print(var1[-5:])

#ตำแหน่งเริ่ม step ปลายทาง [เริ่ม:ปลายทาง:step]
print(var1[0:11:3])
```

```python
#การเพิ่ม แก้ไข ข้อมูลใน String

#การแทรกข้อมูลใน String
var1 = 'Hello World'
print('Update String -> ',var1[:5] + '-' + var1[6:]) #การบวก เรียกอีกอย่างว่า Concatenate การเชื่อมข้อความ
```

## สรุปตัวดำเนินการเกี่ยวกับ String (String special operators)

Operand ตัวถูกดำเนินการ เช่น ตัวแปร ตัวเลข

Operetor ตัวดำเนินการ เช่น +, -, *, /, %

เช่น x + y

Operator คือ +

Operand x,y

Operator สำหรับ String

[ตัวเลขตำแหน่ง] คือ การเข้าถึงตำแหน่งด้วยการระบุตัวเลขตำแหน่ง

เช่น x = "Hello"

ถ้าต้องการเข้าถึงตัว e ต้องใช้ x[1]

[:] คือการเข้าถึงตำแหน่งด้วยการระบุตำแหน่งเริ่มต้น ตัวหลังคือจำนวนทั้งหมดที่ต้องการ

เช่น x = "Hello"
ถ้าต้องการเข้าถึง ell ต้องใช้ x[1:4]

เพราะ 1 คือตำแหน่งของ e และ 4 คือจำนวนตั้งแต่ H ถึง l ตัวที่สอง นับได้ 4 ตัว
