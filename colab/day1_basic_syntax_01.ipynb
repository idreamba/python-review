***Indent (เว้นวรรค)***

```python
x = 3
y = 3
if(x < y):
    print("x น้อยกว่า y")
elif(x == y):
    print("x เท่ากับ y")
else:
    print("x มากกว่า y")
```

***Variable (ตัวแปร)***
1. Int เป็นเลขจำนวนเต็ม เก็บ 2ไบต์ (16บิต) 	0 - 65,535 เช่น 194
2. Long เป็นเลขจำนวนเต็ม 4ไบต์ 0 - 294,967,295
3. Float เป็นเลขทศนิยม เช่น 3.21
4. String เป็นข้อความ เช่น 'Dream', "Dream"
5. Boolean เป็นค่าทางตรรกะ เช่น True, False
6. Complex หรือ Imaginary เป็นตัวเลขเชิงซ้อน 5+4j
