```python
#single quote
say = 'Hello'
sentence = "This is a book"
paragraph = """triple quote ไม่ใช่เพียงแค่ทำ 
comment เท่านั้น 
แต่สามารถใข้เชื่อม string 
แบบหลายบรรทัดได้ด้วย"""
print(say)
print(sentence)
print(paragraph)
```
