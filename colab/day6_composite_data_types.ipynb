## ข้อมูล ทัพเพิล (Tuples)

มีลักษณะโครงสร้างคล้าย List เก็บข้อมูลต่างชนิดกันได้เหมือนกัน แต่ เก็บข้อมูลที่เป็นค่าคงที่ และไม่เปลี่ยน ไม่สามารถเพิ่มหรือลบได้โดยตรง การเข้าถึงข้อมูลทำได้เหมือน List แต่จะอยู่ในเครื่องหมาย ( )



```python
pupprint = ['Nakrit','Aree','36390',13,'0642108180'] #list
x = ('Nakrit','Aree','36390',13,'0642108180') #tuple

print(pupprint[0])
print(x[0])

#List สามารถเปลี่ยนค่าได้
pupprint[4] = 15
print(pupprint[4])
print(len(pupprint))

#Tuple ไม่สามารถเปลี่ยนค่าได้
x[4] = 15
print(x[4])

```

## ชนิดข้อมูล ดิกชันนารี (Dictionary)

คล้ายกับ List เก็บข้อมูลต่างชนิดกันได้ เก็บได้หลายค่า แต่สิ่งที่แตกต่างกัน คือ จะมีการเก็บข้อมูลเป็นคู่ คือมี Key กับ Value

ที่สัมพันธ์กันเสมอ และใช้สัญลัก { } ในการเก็บข้อมูลในนั้น

```python
list_ = ['Nakrit','Aree','36390',13,'0642108180'] #list
tuple_ = ('Nakrit','Aree','36390',13,'0642108180') #tuple

print(list_[0])

#firstname = 'Nakrit'
dictionary = {'firstname':'Nakrit','lastname':'Aree','id':'36390','age':13,'tel':'0642108180'} #dictionary
print(dictionary['firstname'], dictionary['lastname'])
```

## ชนิดข้อมูล เซต (Set)

คล้ายกับ List แต่จะใช้สัญลักษณ์ {x1,x2,x3}

ในการประกาศแบบเซต จะใช้เป็น

x = set([1,2,3,4,5])

```python
# การใช้เซต
# 1.มันจะตัดข้อมูลซ้ำออกให้อัตโนมัติ
# 2.เมื่อประกาศเซต ต้องระวังตำแหน่งที่มันสร้างให้ใหม่

setA = set(['Banana','Orange','Mango'])
setB = set(['Banana','Orange','Mango','Apple','Durian'])

print('setA: ',setA)
print('setB: ',setB)
```

```python
# loop ข้อมูลใน set ออกมา
setA = set(['Banana','Orange','Mango'])
setB = set(['Banana','Orange','Mango','Apple','Durian'])

#ดึงข้อมูลมาใช้
listdat = []
for id in setA:
    listdat.append(id)

print(listdat)
print(listdat[0])
```

```python
# ฟังชันดำเนินการในเซ็ต
setA = set(['Banana','Orange','Mango'])
setB = set(['Banana','Orange','Mango','Apple','Durian'])

# add() เพิ่มได้ทีละ 1 ตัว
setA.add('Grape')
setA.add('Papaya')
setA.add('Apple')
setA

# update() จะอัพเดทข้อมูลที่เป็นชุด เพิ่มเข้าไปรวมกับของเก่า
setA.update(['Mango','Apple','Durian'])
setA

# pop()
x = setA.pop()
x

setA

# romove ใส่ชื่อตรงๆ ที่ต้องการลบ ไม่ใช่ตำแหน่ง
setA.remove('Banana')
setA

# clear() คือ ลบเซตออกทั้งหมด
setA.clear()
setA
```

```python
# operator การดำเนินการในเซ็ต
setA = set(['Banana','Orange','Mango','Apple'])
setB = set(['Banana','Orange','Mango','Apple','Durian'])

# Intersection คือ การเลือกข้อมูลที่เหมือนfc                                                                    vfกันทั้ง 2 เซต มาจัดกลุ่มใหม่
setA & setB

# Union คือ การรวมข้อมูลทุกอย่างที่มีอยู่ในเซ็ต A และ B โดยไม่เอาข้อมูลซ้ำเข้ามารวมด้วย
setA | setB

# Complement (Diff) คือ การเอาข้อมูลที่ไม่เหมือนกัน ที่อยู่ในเซตซ้ายมือเท่านั้น
setA - setB

# Symetric Diff คือ การเลือกข้อมูลที่ทั้ง A และ B ไม่มีเหมือนกัน
setA ^ setB
```
