## 1. ชนิดข้อมูลพื้นฐาน (Basic data types)
แบ่งออกเป็น 2 กลุ่มย่อย คือตัวเลข (Numeric) กับ สายอักษร (String)

**Character** คือ ตัวอักษรเดียว เช่น A, B, C

**String** คือ สายอักษร เอา character มาเรียงต่อกัน เช่น Komkrit

1.1 ชนิดข้อมูลตัวเลข (Numeric) แบ่งออกได้ 2 กลุ่ม คือ 

เลขจำนวนเต็ม (Integer) เช่น -1, 0, 10 ,...

เลขทศนิยม (Float) เช่น 0.11, 1.2234

````python
#Integer
x = 10
print(type(x))

#Float
y = 3.14
print(type(y))
````

````python
# การเอา Interger + Float จะได้ค่า Float ออกมาเสมอ
x = 10
y = 1.12
z = x + y
print(z)
````

## ชนิดข้อมูลแบบ Boolean (จำนวนตรรกยะ)

ค่าจะถูกเก็บเป็นลอจิก 2 ค่า คือ จริง (True) และ เท็จ (False) 

โดยชนิดข้อมูลนี้จะให้ค่าเป็น Integer คือ True = 1, False = 0

````python
# operator +,-,*,/ ทางคำนวนตัวเลข
# operator and, or, not (!) ทางตรรกศาสตร์
# operator : and

# Truth Table and (T = True, F = False)
# -------------------------------------------------------------
#   X   |   Y   |   and     |   or  |   not (and)   |   not (or)
# -----------------------------------------------------------
#   T   |   T   |    T      |   T   |    F          |   F
#   T   |   F   |    F      |   T   |    T          |   F
#   F   |   T   |    F      |   T   |    T          |   F
#   F   |   F   |    F      |   F   |    T          |   T

x = True
y = True
z = not(x and y)
print(z)
````

````python
# and
x = 10
y = 10

# operator เปรียบเทียบ ==, >, <, >=, <=
if(x >= y):
    print('เท่ากัน')
else:
    print('False')
````

## การแปลงค่าตัวแปร (Forcing a number type)

เราสามารถแปลงค่าไปมาระหว่างตัวแปรต่างชนิดกันได้ โดยการทำ forcing

ตาราง ASCII -> https://www.asciitable.com/


````python
x = 65
print(type(x))

y = '65'
print(type(y))

z = int(y) + 1
print(z)
````
