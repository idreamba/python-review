ให้ทำสูตรคูณแม่ 2 ด้วยคำสั่ง for loop





```python
#สูตรคุณแม่ 2

for i in range(1,13):
    print('2 x',i,'=',2*i)
```

```python
#ผลคูณของจำนวนใดๆ ที่ใส่ในช่อง input()

x = input('ใส่ตัวเลข ') #string
y = int(x) # forcing value เป็นการเปลี่ยนค่าชนิดข้อมูลให้กับตัวแปร

for i in range(1,13):
    print(y,'x',i,'=',y*i)
```

## Nested for ลูปซ้อนในลูป





```python
#Nested for

# for x in range(1,11):
#     #print('x',x)
#     for y in range(1,11):
#         print('x',x,'y',y)

# *
# **
# ***
# ****
# *****

for x in range(1):
    star = '*'
    for y in range(x,5):
        print(star + '*'*y)
```

```python
for x in range(1):
    #print('*')
    star = '*'
    for y in range(x,10):
        print(star + '-'*y)
```

```python
# X * y

for x in range(1,5):
    for y in range(1,5):
        print('x'+str(x),'* y'+str(y)+' =',x*y)
```

## ทบทวน While loop

```python
i = 0
while(i<10):
    print(i)
    i = i + 1
```

```python
while(True):
    x = int(input('ใส่ตัวเลข ถ้าใส่เลข 9 จะออกจากโปรแกรม '))
    if(x == 9):
        print('ออกจากโปรแกรม')
        break
    else:
        print('ตัวเลขที่ใส่คือ ',x)

```

```python
# while else
i = 0;
while(i<10):
    print(i)
    #i = i + 1;
    i += 1; #ความหมายเหมือนกับ i = i + 1
else:
    print('ค่าจาก else:',i)
```
